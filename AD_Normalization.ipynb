```python
import pandas as pd
data = pd.read_csv('data.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
data['class'] = data['class'].map({'P':1, 'H':0})
```

```python
X = data.drop(columns=['ID','class'], axis=1)
```

```python
y = data['class']
```

```python
X_df= X
```

```python
X = X.to_numpy()
```

```python
minmax = MinMaxScaler()
x1 = minmax.fit_transform(X)


standard = StandardScaler()
x2 = standard.fit_transform(X)


robust = RobustScaler()
x3 = robust.fit_transform(X)
```

```python
minmaxdf= pd.DataFrame(x1, columns=X_df.columns)
minmaxdf.head()
```

```python
standarddf= pd.DataFrame(x2, columns=X_df.columns)
standarddf.head()
```

```python
robustdf= pd.DataFrame(x3, columns=X_df.columns)
robustdf.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X_df.columns
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols = 4, figsize =(20, 5))
ax1.set_title('Before Scaling')

sns.kdeplot(X_df['air_time1'], ax = ax1, color ='r')
sns.kdeplot(X_df['disp_index1'], ax = ax1, color ='b')
ax2.set_title('After Robust Scaling')


sns.kdeplot(robustdf['air_time1'], ax = ax2, color ='red')
sns.kdeplot(robustdf['disp_index1'], ax = ax2, color ='blue')
ax3.set_title('After Standard Scaling')

sns.kdeplot(standarddf['air_time1'], ax = ax3, color ='black')
sns.kdeplot(standarddf['disp_index1'], ax = ax3, color ='g')
ax4.set_title('After Min-Max Scaling')

sns.kdeplot(minmaxdf['air_time1'], ax = ax4, color ='black')
sns.kdeplot(minmaxdf['disp_index1'], ax = ax4, color ='g')
plt.show()
plt.savefig('n1.jpeg')
```
