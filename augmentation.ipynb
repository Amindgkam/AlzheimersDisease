```python
import pandas as pd
import numpy as np
data = pd.read_csv('data.csv')
```

```python
data['class'] = data['class'].map({'P':1, 'H': 0})
```

```python
a = data.columns
len(a)
```

```python
df =data.drop(columns=['ID'], axis=1)
```

```python
col_means = df.mean(axis=0)
col_stds = df.std(axis=0)
```

```python
col_means
```

```python
n_new_data = 400
```

```python
new_data = pd.DataFrame(columns=df.columns, index=range(n_new_data))
```

```python
new_data
```

```python
for i in range(n_new_data):
    noise = np.random.normal(0, col_stds, size=df.shape[1])
    new_row = col_means + noise
    new_data.iloc[i] = new_row
```

```python
new_data.drop(columns=['class'], axis=1)
```

```python
augmented_data = pd.concat([df, new_data])
```

```python
augmented_data.head()
```

```python
original_labels = df['class'].to_numpy()
label_counts = np.unique(original_labels, return_counts=True)
class_probs = label_counts[1] / np.sum(label_counts)
```

```python
class_probs = np.array([class_probs[i] if i < len(class_probs) else class_probs[-1] for i in range(len(label_counts[0]))])
```

```python
class_probs /= np.sum(class_probs)
```

```python
new_labels = np.random.choice(label_counts[0], size = n_new_data, p = class_probs)
```

```python
new_labels
```

```python
original_labels
```

```python
y = np.concatenate([original_labels, new_labels])
```

```python
y
```

```python
len(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(augmented_data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.15, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.regularizers import L1, L2
```

```python
model = Sequential()
model.add(Dense(1024, activation='relu', input_dim = X_train.shape[1]))
model.add(Dense(1024, activation='relu', ))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=10, validation_data=(X_test, y_test), validation_batch_size=5)
```
